```python
import pandas as pd
import numpy as np

df = pd.read_csv("../data/processed/netflix_cleaned.csv")
df.head()
```

```python
# Create empty columns
df["duration_minutes"] = np.nan
df["seasons"] = np.nan

# If type = Movie → extract minutes
movie_mask = df["type"] == "Movie"
df.loc[movie_mask, "duration_minutes"] = (
    df.loc[movie_mask, "duration"]
    .astype(str)
    .str.extract(r"(\d+)")
    .astype(float)
)

# If type = TV Show → extract seasons
tv_mask = df["type"] == "TV Show"
df.loc[tv_mask, "seasons"] = (
    df.loc[tv_mask, "duration"]
    .astype(str)
    .str.extract(r"(\d+)")
    .astype(float)
)

df[["type", "duration", "duration_minutes", "seasons"]].head(20)
```

```python
import pandas as pd

INPUT_PATH = "../data/processed/netflix_processed.csv"
OUTPUT_PATH = "../data/processed/netflix_processed.csv"

df = pd.read_csv(INPUT_PATH)

# Movies → minutes
df["duration_minutes"] = None
movie_mask = df["type"] == "Movie"
df.loc[movie_mask, "duration_minutes"] = (
    df.loc[movie_mask, "duration"]
    .astype(str)
    .str.extract(r"(\d+)")
    .astype(float)
)

# TV Shows → seasons
df["seasons"] = None
tv_mask = df["type"] == "TV Show"
df.loc[tv_mask, "seasons"] = (
    df.loc[tv_mask, "duration"]
    .astype(str)
    .str.extract(r"(\d+)")
    .astype(float)
)

df.to_csv(OUTPUT_PATH, index=False)

print("✅ Saved successfully!")
print(df.columns)
```

## 🔵 Univariate Analysis

```python
# Distribution of Content Type
type_counts = df["type"].value_counts()
type_counts
```

```python
import plotly.express as px
```

```python
type_counts_df = df["type"].value_counts().reset_index()
type_counts_df.columns = ["Type", "Count"]

fig = px.bar(
    type_counts_df,
    x="Type",
    y="Count",
    text="Count",
    title="Distribution of Movies vs TV Shows"
)

fig.show()
```

```python
rating_counts_df = df["rating"].value_counts().reset_index()
rating_counts_df.columns = ["Rating", "Count"]

fig = px.bar(
    rating_counts_df,
    x="Rating",
    y="Count",
    text="Count",
    title="Distribution of Ratings"
)

fig.show()
```

```python
year_counts_df = df["release_year"].value_counts().sort_index().reset_index()
year_counts_df.columns = ["Year", "Count"]

fig = px.line(
    year_counts_df,
    x="Year",
    y="Count",
    markers=True,
    title="Content Release Trend Over Years"
)

fig.show()
```

COUNTRY DISTRIBUTION


```python
import plotly.express as px

country_counts = df["country"].value_counts().head(10).reset_index()
country_counts.columns = ["Country", "Count"]

fig = px.bar(
    country_counts,
    x="Country",
    y="Count",
    text="Count",
    title="Top 10 Countries by Number of Titles"
)

fig.show()
```

## 🟠 Bivariate Analysis


```python
type_rating = pd.crosstab(df["type"], df["rating"])
type_rating
```

1️⃣ Type vs Rating

```python
fig = px.imshow(
    type_rating,
    text_auto=True,
    aspect="auto",
    title="Type vs Rating Relationship"
)

fig.show()
```

### Insight:

Movies have a wider variety of rating categories.
TV Shows are mostly concentrated in TV-MA and TV-14 ratings,
indicating strong focus on mature and teen audiences.

```python
type_year = df.groupby(["release_year", "type"]).size().reset_index(name="Count")

fig = px.line(
    type_year,
    x="release_year",
    y="Count",
    color="type",
    title="Movies vs TV Shows Over Time"
)

fig.show()
```

### Insight:

Both Movies and TV Shows increased significantly after 2015.
However, TV Shows show faster growth in recent years.

3️⃣ Duration vs Type

```python
# Extract numeric part from duration column
df["duration_numeric"] = df["duration"].str.extract(r"(\d+)").astype(float)

df.head()
```

```python
df_movies = df[df["type"] == "Movie"]

fig = px.histogram(
    df_movies,
    x="duration_numeric",
    nbins=30,
    title="Distribution of Movie Duration"
)

fig.show()
```

- The dataset contains more Movies than TV Shows.

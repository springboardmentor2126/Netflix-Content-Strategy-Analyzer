```python
# Netflix Data Cleaning code

import pandas as pd#Import Libraries
import matplotlib.pyplot as plt
data = pd.read_csv("../data/raw/netflix_titles.csv")#Load the Dataset
data.info()# Initial Data Inspection
```

```python
data.isnull().sum()
data.drop_duplicates(inplace=True)#Remove Duplicate Records
data
```

```python
print("Duplicates Removed")
```

```python
data.dropna(subset=['date_added'], inplace=True)
data
```

```python
data=data.fillna({
    'director':'No data',
    'cast':'No cast',
    'country':'Missing Data',
    'rating':'No Rating'
})
```

```python
data['date_added'] = pd.to_datetime(data['date_added'], errors='coerce')
data.dropna(subset=['date_added'], inplace=True)
data
```

```python
data['year_added']=data['date_added'].dt.year 
data['month_added']=data['date_added'].dt.month
data
```

```python
text_cols=['title','director','cast','country','rating','type']
for col in text_cols:
    data[col]=data[col].str.strip()#Clean Text Columns (Remove Extra Spaces)
```

```python
data['country'].str.split(',').str[0]#Standardize Country Column
```

```python
data[['duration_int', 'duration_type']] = data['duration'].str.split(' ', expand=True)
data['duration_int']=pd.to_numeric(data['duration_int'],errors='coerce')
data.dropna(subset=['duration_int'],inplace=True)#Process and Split Duration Column
data
```

```python
data['content_age']=data['year_added']-data['release_year']
data['release_decade']=(data['release_year']//10)*10 #Feature Engineering
data
```

```python
print("Missing Values:",data.isnull().sum())#Final Data Quality Check
```

```python
data.to_csv("../data/processed/netflix_titles_cleaned.csv")
print("cleaning Completed Successfully")
```

```python
 # ================================
# FEATURE ENGINEERING
# ================================
#Create genre_count column
#Splits 'listed_in' by comma and counts number of genres

data['genre_count'] = data['listed_in'].apply(
    lambda x: len([g.strip() for g in x.split(',')])
)
```

```python
#Create main_genre column
# Extracts first genre from the list
data['main_genre'] = data['listed_in'].apply(
    lambda x: x.split(',')[0].strip()
)
```

```python
#Create main_country column
# Extracts first country if multiple countries exist
data['main_country'] = data['country'].apply(lambda x: x.split(',')[0].strip())
```

```python
#Create content_length_category
# Categorizes movies and TV shows based on duration

def categorize_length(row):
    if row['duration_type'] == 'min':
        if row['duration_int'] < 60:
            return "Short Movie"
        elif row['duration_int'] <= 120:
            return "Medium Movie"
        else:
            return "Long Movie"
    else:
        if row['duration_int'] == 1:
            return "Mini Series"
        else:
            return "Multi Season Series"

data['content_length_category'] = data.apply(categorize_length, axis=1)
```

```python
#Create content_age_category
# Categorizes content based on how old it is

data['content_age_category'] = pd.cut(
    data['content_age'],
    bins=[-1,2,5,10,50],
    labels=['Very New','New','Moderate','Old']
)
```

```python
#Create original_vs_licensed column
# Checks if word "Netflix" exists in title
data['original_vs_licensed'] = data['title'].apply(
    lambda x: "Netflix Original" if "Netflix" in str(x) else "Licensed"
)
```

```python
# Print shape after feature engineering
print("Shape After Feature Engineering:", data.shape)
print(data.columns)
```

```python
data['cast_count'] = data['cast'].apply(
    lambda x: len(str(x).split(',')) if pd.notna(x) else 0
)
```

```python
# =====================================
# MILESTONE 3 - ADVANCED DATA ANALYSIS
# =====================================

import pandas as pd
import matplotlib.pyplot as plt

# Load engineered dataset
data = pd.read_csv("../data/processed/netflix_titles_featured.csv")
```

```python
# -----------------------------
# 1. Group titles by Genre, Rating
# -----------------------------

genre_rating = pd.crosstab(data['main_genre'], data['rating'])

top_genres = data['main_genre'].value_counts().head(8).index

genre_rating.loc[top_genres].plot(
    kind='bar',
    stacked=True,
    figsize=(12,6)
)

plt.title("Content Groups by Genre and Rating")
plt.xlabel("Genre")
plt.ylabel("Number of Titles")
plt.show()
```

```python
# -----------------------------
# 2. Movies vs TV Shows
# -----------------------------

type_counts = data['type'].value_counts()

plt.figure(figsize=(6,6))

plt.pie(
    type_counts.values,
    labels=type_counts.index,
    autopct='%1.1f%%',
    startangle=90
)

plt.title("Movies vs TV Shows Distribution")
plt.show()
```

```python
# Content Growth Over Time

year_counts = data['year_added'].value_counts().sort_index()

plt.figure(figsize=(10,5))

plt.plot(year_counts.index, year_counts.values, marker='o')

plt.title("Netflix Content Growth Over Time")
plt.xlabel("Year Added")
plt.ylabel("Number of Titles")

plt.grid(True)

plt.show()
```

```python
# -----------------------------
# 3. Duration Comparison
# -----------------------------

movie_data = data[data['type']=="Movie"]
tv_data = data[data['type']=="TV Show"]

plt.figure(figsize=(8,5))

plt.hist(movie_data['duration_int'], bins=30, alpha=0.6, label="Movies")
plt.hist(tv_data['duration_int'], bins=10, alpha=0.6, label="TV Shows")

plt.legend()
plt.title("Duration Comparison: Movies vs TV Shows")
plt.xlabel("Duration")
plt.ylabel("Frequency")
plt.show()
```

```python
# -----------------------------
# 4. Country vs Genre Analysis
# -----------------------------

country_genre = pd.crosstab(data['main_country'], data['main_genre'])

top_countries = data['main_country'].value_counts().head(10).index

country_genre.loc[top_countries].plot(
    kind='bar',
    stacked=True,
    figsize=(12,6)
)

plt.title("Country Contribution by Genre")
plt.xlabel("Country")
plt.ylabel("Number of Titles")
plt.show()
```

```python
top_countries = data['main_country'].value_counts().head(10)

plt.figure(figsize=(10,6))

plt.barh(top_countries.index, top_countries.values)

plt.title("Top Countries Producing Netflix Content")
plt.xlabel("Number of Titles")
plt.ylabel("Country")

plt.show()
```

```python
# -----------------------------
# 5. Feature Relationship Analysis
# -----------------------------

numeric_cols = [
    'release_year',
    'year_added',
    'duration_int',
    'genre_count',
    'content_age'
]

corr = data[numeric_cols].corr()

plt.figure(figsize=(8,6))

plt.imshow(corr)

plt.xticks(range(len(corr.columns)), corr.columns, rotation=45)
plt.yticks(range(len(corr.columns)), corr.columns)

plt.colorbar()
plt.title("Feature Correlation Matrix")

plt.show()


# Save for dashboard
data.to_csv("../data/processed/dashboard_data.csv")
```
